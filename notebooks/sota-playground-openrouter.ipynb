# Playground for SOTA Models (OpenRouter)

Setup API key for openrouter: https://openrouter.ai/settings/keys and save en your .env file as OPENROUTER_API_KEY then run:

```python
import os
import json

from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
```

```python
api_key = os.environ["OPENROUTER_API_KEY"]
fpath = "data/openrouter-models.json"
```

```python
!curl -s https://openrouter.ai/api/v1/models \
  -H "Authorization: Bearer $OPENROUTER_API_KEY" | jq '.' > "data/openrouter-models.json"
```

```python
import json

with open(fpath, "r") as f:
    data = json.load(f)

models = data["data"]
print(f"{len(models)} models available\n")
```

List available models and their details/pricing so we can choose one to use.

```python
for model in models:
    model_id = model.get("id", "N/A")
    name = model.get("name", "No name")
    desc = model.get("description", "").strip().split("\n")[0]
    pricing = model.get("pricing", {})
    prompt_price = float(pricing.get("prompt", "0"))
    completion_price = float(pricing.get("completion", "0"))

    # Use ‚úÖ if free, üí∞ if paid
    is_free = prompt_price == 0 and completion_price == 0
    icon = "‚úÖ" if is_free else "üí∞"

    print(f"{icon} {model_id} - {name}")
    print(f"   ‚Ü≥ {desc}")
    print(f"   ‚Ü≥ Prompt: {prompt_price:.10f}, Completion: {completion_price:.10f}\n")
```

```python
client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=api_key,
)
```

```python
completion = client.chat.completions.create(
    # extra_headers={
    #     "HTTP-Referer": "<YOUR_SITE_URL>",  # Optional. Site URL for rankings on openrouter.ai.
    #     "X-Title": "<YOUR_SITE_NAME>",  # Optional. Site title for rankings on openrouter.ai.
    # },
    model="deepseek/deepseek-r1-0528-qwen3-8b",
    messages=[{"role": "user", "content": "What is the meaning of life?"}],
)

print(completion.choices[0].message.content)
```
